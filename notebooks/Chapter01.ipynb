```python
import numpy
import random

from matplotlib import pyplot
```

```python
"""Exercise 1.2

Sketching a regression model and data: Figure 1.1b shows data corresponding to the fitted line
y = 46:3 + 3:0x with residual standard deviation 3.9, and values of x ranging roughly from 0 to
4%.
(a) Sketch hypothetical data with the same range of x but corresponding to the line y = 30 + 10x
with residual standard deviation 3.9.
(b) Sketch hypothetical data with the same range of x but corresponding to the line y = 30 + 10x
with residual standard deviation 10.
"""
```

```python
# Eyeballed from Fig 1.1b
XS = (-0.5, 0.07, 0.4,
      0.95, 1.05, 1.06, 1.1, 
      1.7,  2.2, 2.21, 2.25, 2.9, 3.05,
      3.8, 3.95, 4.2)

Y_MEAN = tuple(30 + 10 * xi for xi in xs)
Y_RAW_DELTA = (-0.1, -0.1, -0.2,
               0.2, 0.4, -0.1, 0.8,
               -0.1, 0.2, -0.4, -0.8, -0.5, 0.4,
               0.6, 0.2, 0.3)

# Manually wiggle this up and down till the number matches
# 9.5 gives resid of 3.84
# 24.9 gives resid of 10.07
MULTIPLIER = 24.9

ys = tuple(mu + MULTIPLIER * delta for (mu, delta) in zip(Y_MEAN, Y_RAW_DELTA))

def resid_std(ys):
    return numpy.std([yi - y_mu_i for (yi, y_mu_i) in zip(ys, Y_MEAN)])

print(MULTIPLIER, ":", resid_std(ys))

      
```

```python
y3 = tuple(mu + 9.51 * delta for (mu, delta) in zip(Y_MEAN, Y_RAW_DELTA))
y10 = tuple(mu + 24.9 * delta for (mu, delta) in zip(Y_MEAN, Y_RAW_DELTA))

fig = pyplot.figure()
ax = fig.gca()

ax.set_xlabel("Average recent growth in personal income (pct'age points)")
ax.set_ylabel("Incumbent party's vote share (pct'age points)")
ax.plot(XS, Y_MEAN, 'k-')
ax.plot(XS, y3, 'b.', label="Resid. std. dev: 3.84")
ax.plot(XS, y10, 'r.', label="Resid. std. dev: 10.07")
ax.legend()
```
