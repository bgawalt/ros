```python
import numpy

from matplotlib import pyplot
from scipy import stats
```

```python
def make_figure_and_axes(draw_yaxis=True) -> (pyplot.Figure, pyplot.Axes):
    fig = pyplot.figure(figsize=(7, 5))
    ax = fig.gca()
    ax.axhline(color='k')
    if draw_yaxis:
        ax.axvline(color='k')
    for side in ('top', 'right', 'bottom', 'left'):
        ax.spines[side].set_visible(False)
    ax.grid()
    return fig, ax
```

# Exercise 3.3, Probability distributions

> Using R, graph probability densities for the normal distribution, plotting
> several different curves corresponding to different choices of mean and
> standard deviation parameters.

```python
xs = numpy.arange(-10, 10, 0.05)

n1 = stats.norm(0, 1).pdf(xs)
n2 = stats.norm(-3, 4).pdf(xs)
n3 = stats.norm(6, 0.6).pdf(xs)

_, ax = make_figure_and_axes()
ax.plot(xs, n1, label="N(0, 1)")
ax.plot(xs, n2, label="N(-3, 4)")
ax.plot(xs, n3, label="N(6, 0.6)")
ax.set_xlim(-10, 10)
ax.legend()
```

# Exercise 3.4, Probability distributions

> Using a bar plot in R, graph the Poisson distribution with parameter 3.5.

```python
xs = numpy.arange(0, 12)
pois_pmf = stats.poisson.pmf(xs, 3.5)

_, ax = make_figure_and_axes(draw_yaxis=False)
ax.bar(xs, pois_pmf, zorder=10)
ax.set_xticks(xs)
ax.set_title("Poisson distribution, mu = 3.5         ", fontsize=16)
ax.set_ylabel("Probability", fontsize=14)
ax.set_xlabel("Count", fontsize=14)
```

# 3.5 Probability distributions

> Using a bar plot in R, graph the binomial distribution with $n = 20$ and
> $p = 0.3$.

```python
xs = numpy.arange(0, 21)
bin_pmf = stats.binom.pmf(xs, n=20, p=0.3)

_, ax = make_figure_and_axes(draw_yaxis=False)
ax.bar(xs, bin_pmf, zorder=10, color='purple')
ax.set_xticks(xs)
ax.set_title("Binomial distribution, n = 20, p = 0.3        ", fontsize=16)
ax.set_ylabel("Probability", fontsize=14)
ax.set_xlabel("Count", fontsize=14)
```

# 3.6 Linear transformations

> A test is graded from 0 to 50, with an average score of 35 and a standard
> deviation of 10. For comparison to other tests, it would be convenient to
> rescale to a mean of 100 and standard deviation of 15.
> 
> (a) Labeling the original test scores as $x$ and the desired rescaled test
> score as $y$, come up with a linear transformation, that is, values of $a$ and
> $b$ so that the rescaled scores $y = a + bx$ have a mean of 100 and a standard
> deviation of 15.
>
> (b) What is the range of possible values of this rescaled score y?
>
> (c) Plot the line showing $y$ vs. $x$.

```python
def transform_score(orig_score: float) -> float:
    return 47.5 + 1.5*orig_score
    
xs = [0, 50]
ys = [transform_score(xi) for xi in xs]

_, ax = make_figure_and_axes()
ax.plot(xs, ys, 'b.-', markersize=19, label=None)
ax.scatter([35,], [transform_score(35)], label='Mean Score',
          color='red', zorder=100, edgecolors='blue', s=57)
ax.scatter([25, 45], [transform_score(25), transform_score(45)], 
          label='Mean +/- one std.', color='orange', zorder=100,
          marker='s', edgecolors='blue', s=57)
ax.set_ylabel("Transformed test score", fontsize=14)
ax.set_xlabel("Original test score", fontsize=14)
ax.set_ylim(0, 130)
ax.legend()
```
